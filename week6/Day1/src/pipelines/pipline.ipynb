```python
import pandas as pd
import numpy as np
import os
```

```python
curr_dir = os.getcwd() 
rawDataPath = os.path.join(curr_dir, '..', 'data', 'raw', 'data.csv')
PROCESSED_PATH = os.path.join(curr_dir, '..', 'data', 'processed', 'processed_data.csv')

DF = pd.read_csv(rawDataPath)
```

```python
DF.head()
```

```python
DF.columns
```

```python
DF.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True, errors="ignore")

# handle missing values
# Age -> median
if "Age" in DF.columns:
    DF["Age"].fillna(DF["Age"].median(), inplace=True)

# Embarked -> mode
if "Embarked" in DF.columns:
    DF["Embarked"].fillna(DF["Embarked"].mode()[0], inplace=True)

# remove duplicates
DF.drop_duplicates(inplace=True)

# handle outliers (Fare)
if "Fare" in DF.columns:
    Q1 = DF["Fare"].quantile(0.25)
    Q3 = DF["Fare"].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    DF["Fare"] = np.clip(DF["Fare"], lower, upper)

# encoding made = 0, female = 1
if "Sex" in DF.columns:
    DF["Sex"] = DF["Sex"].map({"male": 0, "female": 1})

# Embarked -> one-hot
if "Embarked" in DF.columns:
    DF = pd.get_dummies(DF, columns=["Embarked"], drop_first=True)

# save final cleaned data
DF.to_csv(PROCESSED_PATH, index=False)
DF.shape
```
