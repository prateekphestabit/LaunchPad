```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/prateek/Prateek/LaunchPad/week6/Day1/src/data/processed/final.csv')
# df = pd.read_csv('/home/prateek/Prateek/LaunchPad/week6/Day1/src/data/raw/data.csv')

#co-realtion matrix
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr[['Survived']].sort_values(by='Survived', ascending=False),
    annot=True,
    cmap='coolwarm'
)

plt.title("Feature Correlation with Survival")
plt.show()
```

```python
df.info()
```

```python
# Feature Distributions
numeric_cols = df.select_dtypes(include=["int64"]).columns

for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

```

```python
# Target Distribution
plt.figure()
ax = sns.countplot(x="Survived", data=df)
plt.title("Target Distribution (Survived)")

# Add percentages on bars
total = len(df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height,
            '{:1.1f}%'.format(100*height/total),
            ha="center")

plt.show()
```

```python
#missing values heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=True, cmap="viridis", yticklabels=False)
plt.title("Missing Values Heatmap")
plt.xlabel("Features")
plt.tight_layout()
plt.show()
```
